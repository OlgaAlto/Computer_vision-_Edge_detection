```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
```

```python
# Load the image in gray scale ("0")
img = cv2.imread('butterfly.jpg',0)
```

```python
# Open the image
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Contours
edges=cv2.Canny(img,90,200)
cv2.imshow('butterfly_result',edges)
cv2.waitKey(4000)
```

```python
fig, ax = plt.subplots(figsize=(9, 4)), plt.imshow(img)
plt.title('Original Image')

fig, ax = plt.subplots(figsize=(9, 4)), plt.imshow(edges)
plt.title('Edge Image')
```
